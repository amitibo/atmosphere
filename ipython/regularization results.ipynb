```python
from mayavi.api import OffScreenEngine
from mayavi.modules.api import Outline, IsoSurface
from mayavi.sources.api import ArraySource
import mayavi.mlab as mlab
import amitibo
import scipy.io as sio
```

```python
def createSurfaceImg(vals, number_of_contours=10, opacity=0.4, azimuth=0, eleveation=-90):
    
    #
    # Create the MayaVi offscreen engine and scene.
    #
    e = OffScreenEngine()
    e.start()
    win = e.new_scene()
    
    #
    # Add the source
    #
    src = ArraySource()
    src.scalar_data = vals
    e.add_source(src)
    
    #
    # Create the IsoSurfaces
    #
    iso = IsoSurface(compute_normals=False)
    iso.contour.number_of_contours = number_of_contours
    iso.contour.auto_contours = True
    iso.actor.property.opacity = opacity
    e.add_module(iso)

    #
    # Add the outline
    #
    e.add_module(Outline())
    
    #
    # Add the colorbar
    #
    mmsclut = iso.module_manager.scalar_lut_manager
    mmsclut.show_scalar_bar = True
    mmsclutsc = mmsclut.scalar_bar
    mmsclutsc.orientation = "vertical"    # or "horizontal"
    mmsclutsc.width = 0.1
    mmsclutsc.height = 0.8
    mmsclutsc.position = (0.01, 0.15)     # color bar located to the left of the rendering window
    mmsclut.number_of_labels = 10
    mmsclut.number_of_colors = 64
    mmsclut.data_name = "My Label"

    #
    # Create the image
    #
    cam = win.scene.camera
    cam.elevation(eleveation)
    cam.azimuth(azimuth)
    cam.parallel_projection = True
    win.scene.save('data/offscreen.png', size=(1200, 1200))
    
    return imread('data/offscreen.png')


def processRadiance(path):

    data = sio.loadmat(path)

    true_radiance = data['true']
    estimated_radiance = data['estimated']
    objective = data['objective'].ravel()
    
    true_img = createSurfaceImg(true_radiance)
    estim_img = createSurfaceImg(estimated_radiance)
    
    figure(figsize=(24, 8))
    subplot(131)
    imshow(true_img)
    title('True Density')
    subplot(132)
    imshow(estim_img)
    title('Estimated Density')
    subplot(133)
    plot(log(objective))
    title('Log Objective')
    

def showRadiance(path):

    data = sio.loadmat(path)

    Y, X, Z = np.mgrid[0:50:10j, 0:50:10j, 0:100:20j]
    
    true_radiance = data['true']
    estimated_radiance = data['estimated']
    objective = data['objective'].ravel()
    
    print true_radiance.shape
    print Y.shape
    
    true_img = amitibo.viz3DGrid(Y, X, Z, true_radiance)
    true_img = amitibo.viz3DGrid(Y, X, Z, estimated_radiance)
    mlab.show()
```
