### Flux / Irradiance vs Radiance / Intensity

\begin{aligned}
F_0(x)  & = \lim_{\delta A, \delta t \rightarrow 0} \frac{\delta N}{\cos \theta_0 \delta A \times \delta t} \\
\frac{\delta N}{\delta t} & = \cos \theta_0 F_0 \delta A
\end{aligned}

\begin{aligned}
I(\Omega)  & = \lim_{\delta A, \delta t, \delta \Omega \rightarrow 0} \frac{\delta N}{\cos \theta_0 \delta A \times \delta t \times \delta \Omega} \\
\frac{\delta N}{\delta t} & = | n \cdot \Omega | I(\Omega) \delta \Omega \delta A
\end{aligned}

\begin{aligned}
\text{scalar flux: } J(x) & = \int_{4\pi} I (x, \Omega) d\Omega \\
\text{vector flux: } \mathbf{F}(x) & = \int_{4\pi} \Omega I (x, \Omega) d\Omega = 
\left( \begin{array}{c}
F_x \\
F_y \\
F_z \end{array} \right) \\
F_n(x) & = \mathbf{n} \cdot \mathbf{F}(x) 
\end{aligned}

### Scattering Phase Function

\begin{aligned}
p(x, \Omega_0 \rightarrow \Omega)  & = \lim_{\delta F_0, \delta \Omega \rightarrow 0} \frac{\delta F_s}{\delta F_0 \times \delta \Omega}
\end{aligned}

### Integro-Differential Radiative Transfer Equation

\begin{aligned}
\Omega \cdot \nabla I &= - \sigma(x) I(x, \Omega) + S(x, \Omega) + Q(x, \Omega) \\
\end{aligned}

where $\Omega$ is the direction of change, $S$ is the scattering source and $Q$ is the thermal source. This can also be written as:

\begin{aligned}
L I &= Q \\
L &= \Omega \cdot \nabla + \sigma(x) \int_{4\pi} \{\delta(\Omega' - \Omega) - \varpi_0 (x) p(x, \Omega'\cdot\Omega)\} [\cdot] d \Omega'
\end{aligned}

### Integral Radiative Transfer Equation

BC for the domain $M$, a solar beam with flux $F_0$ and incidence direction $\Omega_0$

\begin{aligned}
f(x, \Omega) = F_0 \delta(\Omega - \Omega_0) 1_{\partial M_{sunny(\Omega_0)}}(X_0)
\end{aligned}

Solution where there is no scattering source $S(x, \Omega) \equiv 0$.

\begin{aligned}
I(x, \Omega) = &f(x_{\partial M}(x, \Omega), \Omega)\exp[-\tau(s_{\partial M}(x, \Omega); x, -\Omega)] \\
&+ \int_0^{s_{\partial M}(x, \Omega)}Q(x-\Omega s, \Omega)\exp [-\tau(s;x, -\Omega)]ds
\end{aligned}

In the presence of volume scattering:

\begin{aligned}
I(x, \Omega) = &I_{fQ}(x, \Omega) \\
               &+ \int_0^{s_{\partial M}(x, \Omega)}\sigma_x(x-\Omega s)\exp [-\tau(s;x, -\Omega)] \times \int_{4\pi}p(x-\Omega s, \Omega' \rightarrow \Omega) I(x-\Omega s, \Omega') d \Omega' ds
\end{aligned}

**Explanation:**

The radiance at point $x$ and direction $\Omega$ is:

* $I_{fQ}(x, \Omega)$: the solar beam $f$ and thermal sources $Q$ that emit in direction $\Omega$ along the ray $-\Omega s$ that starts in $x$.
* The scattered sources emited in direction $\Omega$ along the ray $-\Omega s$ that starts in $x$. This includes the light scattered at $x$ itself, in direction $\Omega$.

Converting the double integral to a volume integral by the following equality

\begin{aligned}
dx' &= s^2 ds d \Omega` \\
I(x, \Omega) &= \int_{4\pi}\int_M K_I(x', \Omega' \rightarrow x, \Omega)I(x', \Omega')dx'd\Omega' + I_{fQ}(x, \Omega) \\
K_I(x', \Omega' \rightarrow x, \Omega) &= \sigma_s(x')p(x', \Omega' \rightarrow x, \Omega)\delta\left(\Omega'-\frac{x' - x}{||x' - x||^2}\right)\frac{\exp[-\tau(x', x)]}{||x' - x||^2}
\end{aligned}

It is possible to write as source equation:

\begin{aligned}
S(x, \Omega) &= \int_{4\pi}\int_M K_S(x', \Omega' \rightarrow x, \Omega)S(x', \Omega')dx'd\Omega' + I_{fQ}(x, \Omega) \\
\end{aligned}
