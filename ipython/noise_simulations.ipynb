```python
from __future__ import division
from scipy.misc import lena
img = lena()
```

# Simulation Parameters

### Quantum Efficency

```python
QE = 0.3
```

### Full Well Depth

```python
F = 20000
```

### Bits per Pixel

```python
B = 8
```

### Scale (Electrons to Gray levels)

```python
alpha = 2**B/F
print alpha
```

### Dark Noise Variance

Baer, Richard L. "A model for dark current characterization and simulation." Electronic Imaging 2006. International Society for Optics and Photonics, 2006.

```python
DN_mean = 6
DN_sigma = 2
```

### Exposure Time

```python
t = 10
```

# Simulation

```python
#
# Calculate the number of photons
#
photons = np.random.poisson(lena() * t)
photons[photons>F] = F

#
# Convert to electrons and add the dark noise
# based on Baer, Richard L. "A model for dark current characterization and simulation." Electronic Imaging 2006. International Society for Optics and Photonics, 2006.
#
electrons = (QE*photons.astype(np.float)).astype(np.uint32)
DN_noise = np.random.lognormal(mean=log(DN_mean), sigma=log(DN_sigma), size=electrons.shape).astype(np.uint32)
electrons += DN_noise

#
# Convert to gray level
#
g = electrons.astype(np.float) * alpha

#
# Quantisize
#
qg = g.astype(np.uint32)
qg[qg>2**B] = 2**B
```

# Results

```python
figure()
imshow(photons, cmap='gray')
title('Photons')
figure()
imshow(electrons, cmap='gray')
title('Electrons')
figure()
imshow(g, cmap='gray')
title('Gray levels')
figure()
imshow(g, cmap='gray')
title('Gray levels Quantisized')
```

```python
hist, bins = np.histogram(g, bins=256)
plot(bins[:-1], hist)
```
