```python
from scipy.misc import lena
import scipy.ndimage.filters as filters
```

```python
img = lena()
lap_img = filters.laplace(img)
```

```python
figure(figsize=(16, 8))
subplot(121)
imshow(img, cmap='gray')
title('original')
subplot(122)
imshow(lap_img, cmap='gray')
title('laplacian')
```

### Add Noise

```python
sigma = 20
img_noise = img + sigma * np.random.randn(*img.shape)
img_noise[img_noise<0] = 0
img_noise[img_noise>255] = 255

figure(figsize=(8, 8))
imshow(img_noise, cmap='gray')
title('with noise $\sigma$=%g' % sigma)
```

## Optimization

### Objective and Gradient

```python
Lambda = 2

def objective(x):
    temp = x.reshape(img.shape)
    
    obj = np.linalg.norm(img_noise - temp)**2 + Lambda*np.linalg.norm(filters.laplace(temp))**2
    
    return obj


def gradient(x):
    temp = x.reshape(img.shape)
    
    
    grad = -(img_noise - temp) + 2 * Lambda * filters.laplace(filters.laplace(temp))
    
    return grad.ravel()


objective_graph = []
def intermidiate(x):
    global objective_graph
    
    objective_graph.append(objective(x))
```

### Solver

```python
from scipy.optimize import fmin_l_bfgs_b as solver

x, obj, info = solver(
                func=objective,
                x0=np.zeros(img.size),
                fprime=gradient,
                bounds=[(0, 255)]*img.size,
                maxfun=1000,
                callback=intermidiate
                )

print info['task']
```

```python
figure(figsize=(16,16))
subplot(221)
imshow(img, cmap='gray')
title('original')
subplot(222)
imshow(img_noise, cmap='gray')
title('noised')
subplot(224)
imshow(x.reshape(img.shape), cmap='gray')
title('denoised')
subplot(223)
plot(log(objective_graph))
title('log objective')
```
