### Air Exctinction

```python
from atmotomo import calcHG, L_SUN_RGB, RGB_WAVELENGTH, getResourcePath
from atmotomo import RGB_WAVELENGTH, getResourcePath, getMisrDB, density_clouds1, calcAirMcarats, Mcarats, SOLVER_F3D
import amitibo

atmosphere_params = amitibo.attrClass(
    cartesian_grids=(
        slice(0, 50, 1.0), # Y
        slice(0, 50, 1.0), # X
        slice(0, 10, 0.1)   # H
        ),
    earth_radius=4000,
    air_typical_h=8,
    aerosols_typical_h=2
)

A_air, A_aerosols, Y, X, H, h = density_clouds1(atmosphere_params)
z_coords = H[0, 0, :]*1000

KM_TO_METER = 1000
mcarats_air_exts = calcAirMcarats(z_coords)
single_air_ext = [1.09e-3 * lambda_**-4.05 * A_air[25, 25, :] for lambda_ in RGB_WAVELENGTH]

figure()
for i, c in zip(range(3), ('r', 'g', 'b')):
    plot(z_coords, mcarats_air_exts[i]*KM_TO_METER, c)
    plot(z_coords, single_air_ext[i], c+'+')
title('Air Extinction Coefficients')
```

This graph compares the extinction coefficients of the air as used by MCARaTS and the exctinction coefficients as used in the original version of the single scattering (based on the LEO book).

Note that in the case of the MCARaTS the coefficient is given in [$m^{-1}$] meters while the single uses [$KM^{-1}$]. Therefore it is necessary to multiply the MCARAtS coefficients by 1000.

The small difference in extinction coefficients can be attributed to the difference in wavelengths (not checked).

### Aerosols Extinction

The k parameter (spectral extinction crosssection) is given in [$\mu m^2$]. Multiplying it by $10^{-12}$ gives the crossection in [$m^2$].

The single scattering coefficients is calculated so that the *Mean Free Path* of one of the channels (the one with the highest crosssection) will be equal to **VISIBILITY**.

The mcarats scattering coefficient should be calculated in the same manner just multiplied by the ratio of meters to km.

```python
particle = getMisrDB()['spherical_nonabsorbing_2.80']

VISIBILITY = 100

k_single=np.array(particle['k']) / np.max(np.array(particle['k']))
single_aerosols_ext = k_single[0] * A_aerosols / VISIBILITY
mcarats_aerosol_ext = single_aerosols_ext / KM_TO_METER

figure()
subplot(121)
imshow(mcarats_aerosol_ext[25, :, :]*KM_TO_METER)
colorbar()
title('Mcarats Aerosols Extinction')
      
subplot(122)
imshow(single_aerosols_ext[25, :, :])
title('Single Aerosols Extinction')
colorbar()
```
